```python
#  Step 2a: Fix Employment_Status inconsistencies

# Preview unique values before cleaning
print("Before cleaning:\n", df['Employment_Status'].value_counts(dropna=False))

# Define a cleaning function
def clean_employment_status(x):
    if pd.isna(x):
        return np.nan
    x = str(x).strip().lower()
    if x in ['emp', 'employed']:
        return 'Employed'
    elif x == 'self-employed':
        return 'Self-Employed'
    elif x == 'unemployed':
        return 'Unemployed'
    elif x == 'retired':
        return 'Retired'
    else:
        return x.title()  # fallback: capitalize first letter

# Apply cleaning
df['Employment_Status'] = df['Employment_Status'].apply(clean_employment_status)

# Check results
print("\nAfter cleaning:\n", df['Employment_Status'].value_counts(dropna=False))
```

```python
# ------------------------------------------------------
#  Step 4 (Pre-Analysis): Outlier Detection & Insights
# ------------------------------------------------------
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# List of numeric columns to check
numeric_cols = [
    'Age', 'Income', 'Credit_Score', 'Credit_Utilization',
    'Missed_Payments', 'Loan_Balance', 'Debt_to_Income_Ratio'
]

# 1 --- Use IQR (Interquartile Range) to detect outliers ---
outlier_summary = []
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)][col]
    outlier_summary.append({
        'Feature': col,
        'Lower_Bound': round(lower, 2),
        'Upper_Bound': round(upper, 2),
        'Outlier_Count': len(outliers),
        'Outlier_%': round((len(outliers) / len(df)) * 100, 2)
    })

outlier_df = pd.DataFrame(outlier_summary)
display(outlier_df)

# 2 --- Visualize with boxplots ---
for col in numeric_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col], color="skyblue")
    plt.title(f"Boxplot: {col}")
    plt.tight_layout()
    plt.show()

# 3 --- Identify and review extreme records ---
# For example, Income top/bottom 1%
high_income = df[df['Income'] > df['Income'].quantile(0.99)]
low_income = df[df['Income'] < df['Income'].quantile(0.01)]

print("\nHigh-income outliers:\n", high_income[['Customer_ID','Income','Credit_Score','Delinquent_Account']].head())
print("\nLow-income outliers:\n", low_income[['Customer_ID','Income','Credit_Score','Delinquent_Account']].head())

# Check utilization >1
util_outliers = df[df['Credit_Utilization'] > 1]
print(f"\nCredit Utilization > 1.0 (invalid cases): {util_outliers.shape[0]}")
display(util_outliers[['Customer_ID','Credit_Utilization']])
```

```python
# 1 Drop rows with missing Credit_Score (too few  safe to remove)
# ----------------------------------------------------
print(f"Before dropping missing Credit_Score: {df.shape[0]} rows")
df = df[df['Credit_Score'].notna()]
print(f"After dropping missing Credit_Score: {df.shape[0]} rows")

# ----------------------------------------------------
# 2 Mean imputation for Income
# ----------------------------------------------------
mean_income = df['Income'].mean()
df['Income'].fillna(mean_income, inplace=True)
print(f"\n Mean Income used for imputation: {mean_income:.2f}")

# ----------------------------------------------------
# 3 AI-assisted synthetic data generation for Loan_Balance
# (Generate realistic values per Credit_Card_Type group)
# ----------------------------------------------------
def synthetic_loan_balance(row):
    if pd.isna(row['Loan_Balance']):
        group = df[df['Credit_Card_Type'] == row['Credit_Card_Type']]['Loan_Balance'].dropna()
        if len(group) > 5:
            mean, std = group.mean(), group.std()
            synthetic_value = np.random.normal(mean, std)
            # Keep values within realistic range (>= 0)
            return max(synthetic_value, 0)
        else:
            return df['Loan_Balance'].mean()
    return row['Loan_Balance']

df['Loan_Balance'] = df.apply(synthetic_loan_balance, axis=1)

# ----------------------------------------------------
# 4 Verify missing values after imputation
# ----------------------------------------------------
print("\n Remaining Missing Values:")
print(df.isna().sum())

# ----------------------------------------------------
# 5 Sanity check for distributions after imputation
# ----------------------------------------------------
print("\n Loan_Balance Summary After Synthetic Generation:")
print(df['Loan_Balance'].describe())

print("\n Income Summary After Mean Imputation:")
print(df['Income'].describe())
```

###  Missing Data Analysis

The dataset contained limited missing data in **Credit_Score**, **Income**, and **Loan_Balance**.
Consistent with financial-industry standards and data-governance guidelines, the following treatments were applied to ensure completeness and reliability:

---

####  Credit Score:
- Since less than **1%** of records were missing this critical variable, those rows were **removed**.  
- Dropping them prevents introducing artificial credit scores and maintains data integrity for a key risk indicator.

---

####  Income:
- Missing values (~7.8%) were imputed using the **mean value**, as the variable exhibited a near-normal distribution with no significant outliers.  
- This method preserves the overall financial distribution and average customer profile while ensuring numerical stability.

---

####  Loan Balance:
- Missing values (~5.8%) were replaced using an **AI-assisted synthetic data generation approach**, where values were generated based on the statistical characteristics (mean and standard deviation) of customers with the same `Credit_Card_Type`.  
- This ensures realistic, segment-consistent loan balances while maintaining data variability and fairness.

---

####  Alignment with Industry Standards:
These treatments align with **Basel II/III data-quality expectations** and **credit-risk modeling best practices**, ensuring the dataset remains both **statistically sound** and **ready for predictive modeling** while maintaining realistic financial relationships.

### Key Findings and Risk Indicators

Exploratory analysis and correlation assessment were conducted to identify key financial and behavioral variables influencing account delinquency.
The correlations with the target variable (`Delinquent_Account`) were relatively weak overall, indicating that delinquency risk is likely driven by **multi-factor behavioral interactions** rather than a single dominant feature.

---

####  Correlation Insights:
- The strongest positive relationships with delinquency were observed for **Income (r = 0.044)**, **Credit Utilization (r = 0.036)**, and **Credit Score (r = 0.035)**  although these are weak linear correlations.  
- **Debt-to-Income Ratio (r = 0.035)** and **Age (r = 0.022)** also showed minor positive associations with delinquency, suggesting slightly higher risk among customers with heavier debt loads or older profiles.  
- **Loan Balance (r = 0.007)** exhibited a negligible relationship with delinquency, implying that absolute loan size alone is not a strong predictor of default.  
- **Missed Payments (r = 0.025)** and **Account Tenure (r = 0.040)** showed weak negative correlations, indicating that some customers may miss occasional payments without becoming delinquent, and longer-tenure customers are marginally more stable.

---

####  Behavioral Segmentation Patterns:
- **Employment Status:**  
  - Highest delinquency among *Unemployed* customers (19.3%), followed by *Employed* (16.3%) and *Self-Employed* (16.3%).  
  - *Retired* customers showed the lowest delinquency (11.6%), indicating higher financial stability.

- **Credit Card Type:**  
  - *Business* cardholders had the highest delinquency (21.3%), followed by *Student* cards (17.9%).  
  - Premium cards like *Gold* (14.5%) and *Platinum* (12.0%) showed stronger repayment behavior and lower risk.

---

####  Key Risk Indicators Identified:
1. **Unemployment:** Strongly associated with higher delinquency probability due to income instability.  
2. **Business and Student Cardholders:** Higher default rates, likely driven by cash flow uncertainty or limited credit experience.  
3. **High Credit Utilization:** Mildly correlated with delinquency  customers nearing credit limits face repayment pressure.  
4. **Lower Account Tenure:** Newly opened accounts show slightly higher risk, consistent with limited repayment history.

---

####  Unexpected Findings:
- Higher-income customers displayed marginally higher delinquency rates, which may indicate **over-leverage** or **lifestyle-driven borrowing patterns**.  
- The weak correlation between `Missed_Payments` and `Delinquent_Account` suggests that not all missed payments escalate to delinquency  implying **possible recovery behavior**.

---

####  Summary:
While no single numeric variable strongly predicts delinquency, **segment-based behavioral differences** (such as employment type, credit card category, and utilization behavior) provide valuable signals for **risk modeling** and **customer profiling**.

```python
# 3. Total_Monthly_Delinquencies (from Month_1..Month_6)
month_cols = ['Month_1','Month_2','Month_3','Month_4','Month_5','Month_6']
df['Total_Monthly_Delinquencies'] = df[month_cols].apply(
    lambda row: sum([1 if x == "Late" else 0 for x in row]), axis=1
)

# 4. Avg monthly delinquency
df['Avg_Monthly_Delinquency'] = df['Total_Monthly_Delinquencies'] / 6

# 5. Recent 3-month delinquency
recent_cols = ['Month_4', 'Month_5', 'Month_6']
df['Recent_3_Month_Delinquency'] = df[recent_cols].apply(
    lambda row: sum([1 if x == "Late" else 0 for x in row]), axis=1
)

# 6. Delinquency trend (Month_1  Month_6 progression)
df['Delinquency_Trend'] = df['Recent_3_Month_Delinquency'] - df['Avg_Monthly_Delinquency']

# 7. Income_per_Debt
df['Income_per_Debt'] = df['Income'] / (df['Loan_Balance'] + 1)

# 8. Risk interaction score
df['Risk_Score_Interaction'] = df['Credit_Score'] * (1 - df['Credit_Utilization'])

# 9. Miss Rate
df['Miss_Rate'] = df['Missed_Payments'] / (df['Account_Tenure'] + 1)

# ================================================================
# 3. SELECT ALL NUMERIC FEATURES for correlation + VIF
# ================================================================
numeric_df = df.select_dtypes(include=['int64', 'float64'])
print("\nNumeric feature count:", numeric_df.shape[1])
print(numeric_df.head())

# ================================================================
# 4. CORRELATION MATRIX (ALL NUMERIC FEATURES)
# ================================================================
plt.figure(figsize=(14, 10))
sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')
plt.title("Correlation Matrix: All Numeric Features")
plt.show()

# ================================================================
# 5. VIF CALCULATION for ALL NUMERIC FEATURES
# ================================================================
scaler = StandardScaler()
numeric_scaled = pd.DataFrame(
    scaler.fit_transform(numeric_df),
    columns=numeric_df.columns
)

print("\n========================")
print("Variance Inflation Factor (All Numeric Features)")
print("========================")

vif = pd.DataFrame()
vif["Feature"] = numeric_scaled.columns
vif["VIF"] = [
    variance_inflation_factor(numeric_scaled.values, i)
    for i in range(numeric_scaled.shape[1])
]

print(vif.sort_values(by="VIF", ascending=False))
```

```python
# ============================================
# 1. IMPORTS & FIX RANDOMNESS
# ============================================
import numpy as np
import random
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report
import joblib

np.random.seed(42)
random.seed(42)

# ============================================
# 2. SELECT FEATURES (OLD MODEL FEATURES)
# ============================================
X = df[['Loan_to_Tenure', 'Income', 'Employment_Risk']]
y = df['Delinquent_Account']   # <-- your target

# ============================================
# 3. TRAINTEST SPLIT (STABLE)
# ============================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    stratify=y,
    random_state=42
)

# ============================================
# 4. SCALING
# ============================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# ============================================
# 5. LOGISTIC REGRESSION
# ============================================
log_model = LogisticRegression(class_weight='balanced', random_state=42)
log_model.fit(X_train_scaled, y_train)

log_train_acc = accuracy_score(y_train, log_model.predict(X_train_scaled))
log_test_acc  = accuracy_score(y_test, log_model.predict(X_test_scaled))
log_auc       = roc_auc_score(y_test, log_model.predict_proba(X_test_scaled)[:,1])

print("\n==============================")
print("LOGISTIC REGRESSION RESULTS")
print("==============================")
print(f"Train Accuracy: {log_train_acc:.3f}")
print(f"Test  Accuracy: {log_test_acc:.3f}")
print(f"AUC: {log_auc:.3f}")
print(classification_report(y_test, log_model.predict(X_test_scaled)))

# ============================================
# 6. DECISION TREE (FINAL MODEL)
# ============================================
tree_model = DecisionTreeClassifier(
    class_weight='balanced',
    max_depth=4,
    random_state=42
)

tree_model.fit(X_train_scaled, y_train)

tree_train_acc = accuracy_score(y_train, tree_model.predict(X_train_scaled))
tree_test_acc  = accuracy_score(y_test, tree_model.predict(X_test_scaled))
tree_auc       = roc_auc_score(y_test, tree_model.predict_proba(X_test_scaled)[:,1])

print("\n==============================")
print("DECISION TREE RESULTS")
print("==============================")
print(f"Train Accuracy: {tree_train_acc:.3f}")
print(f"Test  Accuracy: {tree_test_acc:.3f}")
print(f"AUC: {tree_auc:.3f}")
print(classification_report(y_test, tree_model.predict(X_test_scaled)))

# ============================================
# 7. CHECK OVERFITTING
# ============================================
print("\n==============================")
print("OVERFITTING CHECK")
print("==============================")

print(f"LogisticReg  Overfit Gap = {abs(log_train_acc - log_test_acc):.3f}")
print(f"DecisionTree Overfit Gap = {abs(tree_train_acc - tree_test_acc):.3f}")

# ============================================
# 8. SELECT BEST MODEL
# ============================================
if tree_test_acc > log_test_acc:
    final_model = tree_model
    print("\nFinal Model Selected: DECISION TREE")
else:
    final_model = log_model
    print("\nFinal Model Selected: LOGISTIC REGRESSION")

# ============================================
# 9. SAVE FINAL MODEL & SCALER
# ============================================
pipeline = {
    "scaler": scaler,
    "model": final_model
}

joblib.dump(pipeline, "final_decision_tree_pipeline.joblib")
print("\nModel saved as final_decision_tree_pipeline.joblib")
```
